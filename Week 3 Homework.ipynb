• 2) Repeat the exercise fusing the central difference scheme.
Why this scheme is better?

```python
def forward(x,dx):
    return (np.cos(x+dx) - np.cos(x))/dx
def ferror(x,dx):
    return (np.abs(-np.sin(x)-forward(x,dx)))

x =  ([10**-i for i in range(1,17)])
fy = [ferror(np.pi/4,i) for i in x]
plt.loglog(x,fy,basex=10)

def central(x,dx):
    return (np.cos(x+0.5*dx) - np.cos(x-0.5*dx))/(dx)
def cerror(x,dx):
    return np.abs(-np.sin(x)-central(x,dx))
plt.ylabel(' Error ')
plt.xlabel('dx')
plt.title('Error of Forward and Central Differences')
cy = [cerror(np.pi/4,i) for i in x]
plt.loglog(x,cy,basex=10,color ='r')
plt.show()
```

The error for finite differences is $O(h^{n})$ where $h = dx$ and $n$ is the order ( $n = 1$ for foward and $n = 2$ for central).

Optimal is around $10^{-8}$ for foward difference.

Optimal is around $10^{-5}$ for central difference.

Central difference is of order $O(h^{2})$ so we expect it to perform better, and it does having around 3 orders of magnitude less error.

3) Repeat (2) for x=0. Why do the results differ?

```python
def forward(x,dx):
    return (np.cos(x+dx) - np.cos(x))/dx
def ferror(x,dx):
    return (np.abs(-np.sin(x)-forward(x,dx)))

x =  ([10**-i for i in range(1,17)])
fy = [ferror(0,i) for i in x]
plt.loglog(x,fy,basex=10)

def central(x,dx):
    return (np.cos(x+0.5*dx) - np.cos(x-0.5*dx))/(dx)
def cerror(x,dx):
    return np.abs(-np.sin(x)-central(x,dx))

plt.ylabel('Log Error ')
plt.xlabel('dx')
plt.title('Error of Forward and Central Differences')

cy = [cerror(0,i) for i in x]
plt.plot(x,cy,color ='r')
plt.show()
```

Central difference is exact , ie error = 0 for x = 0 and so does not show up in the plot. This is because cos(x) is symmetric around x = 0 and thus for central difference the terms cancel out and thus equal to the exact solution -sin(x) for x = 0.
