```python
%matplotlib inline
import numpy as np

import matplotlib.pyplot as plt
```

```python

N = 5
def f(t, y):
    f=np.zeros(2)
    f[0]=y[1]
    f[1]= (2*t*y[1]-N*(N+1)*y[0])/(1-t*t)
    return f
    
def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t,y)
    k2=h*f(t+0.5*h,y+0.5*k1)
    k3=h*f(t+0.5*h,y+0.5*k2)
    k4=h*f(t+h,y+k3)
    y+=(k1+2.*(k2+k3)+k4)/6.
    return y
```

```python
for dy in np.linspace(0,2.5,5):
    y = [0,dy]
    tend = 1
    tn = 1000
    tp = np.linspace(0,1,tn) 
    yp = [y[0]]
    dyp = [y[1]]
    hp = []
    for i in range(0,tn-1):
        if i ==  tn - 1 :
            h = tp[i] - tp[i-1]
            hp = np.append(hp,h)
        else:
            h = tp[i+1]-tp[i]  
            hp = np.append(hp,h)

        t = tp[i]
        y = rk4(t,h,y,4)
        yp = np.append(yp,y[0])
        dyp = np.append(dyp,y[1])
    plt.plot(tp,yp)
plt.show()
```
